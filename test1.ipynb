```python
for stock in tqdm(tmp['stock'].values):
    if os.path.exists('data/{}.pkl'.format(str(stock))):
        continue
    data = get_price(security=stock, start_date='2023-11-30 00:00:00', end_date='2024-12-06 00:00:00', frequency='daily', fields=['open','close','low','high','volume','money','factor',
        'high_limit','low_limit','avg','pre_close','paused'], skip_paused=False, fq='pre', count=None,round=True)
    data.to_pickle('data/{}.pkl'.format(str(stock)))
```
