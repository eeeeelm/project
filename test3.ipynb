```python
import pandas as pd
import numpy as np
import os
import torch
```

```python
stocks = pd.read_pickle('data/stock_code.pkl')
stocks = stocks['stock'].values
for stock in stocks:
    if os.path.exists('data/' + str(stock) + '.pkl') == False:
        break
    data = pd.read_pickle('data/' + str(stock) + '.pkl')
    data = data[~data.isnull().any(axis=1)]
    data[['open_rate', 'close_rate', 'low_rate', 'high_rate', 'avg']] = data[['open', 'close', 'low', 'high', 'avg']].sub(data['pre_close'], axis=0).div(data['pre_close'], axis=0)
    data = data[['open_rate', 'close_rate', 'low_rate', 'factor', 'high_rate', 'avg', 'close']]
    data.to_pickle('part_data/' + str(stock) + '.pkl')
```
